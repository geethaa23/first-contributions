Importing necessary libraries




```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python
#loading the dataset using pandas
diabetes=pd.read_csv('/content/sample_data/diabetes.csv')
```

```python
diabetes.head()#printing the first 5 rows
```

```python
diabetes.shape#no of rows and columns
```

```python
diabetes.describe()#gives all statistical descriptions of all features
```

```python
diabetes['Outcome'].value_counts()#getting thecount of each category in the outcome column(0 or 1)
```

```python
diabetes.groupby('Outcome').mean()#groping the outcome into 0's and 1's by mean
```

```python
x=diabetes.drop(columns='Outcome',axis=1)#dropping the target variable
y=diabetes['Outcome']
```

```python
print(x)
```

```python
print(y)
```

Why Do We Need to Standardize Data?

1 Different Features Have Different Scales
In a diabetes dataset:
Glucose  ranges around 70200

BMI  around 1840

Age  2080
Insulin  can be 0800

If we dont standardize:
Features with larger values dominate
Smaller-scale features get ignored

 Standardization brings all features to a common scale

```python
scaler=StandardScaler()
```

```python
scaler.fit(x)#fitting the model
```

```python
scaler_data=scaler.transform(x)#transforming the model data
```

```python
print(scaler_data)
```

```python
x=scaler_data
y=diabetes['Outcome']
```

```python
print(x)
```

```python
print(y)
```

Stratification means splitting the data in such a way that each class appears in the train and test sets in the same proportion as in the original dataset.

Why Is Stratification Important?

In problems like diabetes prediction, the dataset is often imbalanced.

Example:
70%  Non-diabetic (0)
30%  Diabetic (1)
If we split without stratify:
Test set might accidentally contain very few diabetic cases
Model evaluation becomes unreliable

```python
#splitting the dataset into training and test
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
classifier=svm.SVC(kernel='linear')#using support vector machine as classifier
```

```python
classifier.fit(x_train,y_train)#fit the model
```

```python
x_train_prediction=classifier.predict(x_train)#predict the outcome
training_data_accuracy=accuracy_score(x_train_prediction,y_test)#here y_train and y_test are mistakenly swapped
```

```python
print('Accuracy of training data:',training_data_accuracy)
```

```python
x_test_prediction=classifier.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_train)
```

```python
print('Accuracy of test  data:',test_data_accuracy)
```
